```python
# importing libraries that will help visualize data
import pandas as pd
import matplotlib.pyplot as plt

# will generate the graph within the notebook
%matplotlib inline 
```

```python
# reading the data file from the local machine into a DataFrame
women_degrees = pd.read_csv('C:/Users/Ivor/Documents/Atkin/Data Analysis/DataQuest/IntermediatePythonAndPandas/\
StoryTellingThroughDataVisualization/percent-bachelors-degrees-women-usa.csv')
```

```python
"""
There are 17 degrees that I will generate line charts for.
I have grouped the degrees into the 3 differet categories:
    * Science, Technology, Engineering and Math (STEM)
    * Liberal Arts
    * Other Category
"""

stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering',\
'Computer Science']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance',\
'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

```python
# creating a grid with figure size of 16 by 20
fig = plt.figure(figsize=(16, 20))

# Generating a 6 row by 3 column grid of subplots
# I am adding the x-axis labels only for the bottom most line charts for better visualization, this is because we are 
# primarily interested in enabling the viewer to quickly get a high level understanding of which degrees are prone to 
# gender imbalance and how has that changed over time. 

# I am also only going to keep the starting and ending y-axis labels which are 0 and 100, in order to reduce clutter
# Removing the y-axis label made it hard to understand which degrees have close to 50-50 gender breakdown
# Therefore I am generating a horizontal line across each graph where the y-axis label 50 would have been
# I will be using the light gray from the Color Blind 10 palette
# http://tableaufriction.blogspot.ro/2012/11/finally-you-can-use-tableau-data-colors.html

# Generating the 1st Column
# Here I generate line charts for both male and female percentages for every degree in the STEM category
for sp in range(0,18,3):
    cat_index = int(sp/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[cat_index]], c='blue', label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[cat_index]], c='orange', label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[cat_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    ax.set_yticks([0,100])
    ax.axhline(50, c=(207/255, 207/255, 207/255), alpha=0.3)
    
    if cat_index == 0:
        ax.text(2005, 80, 'Women')
        ax.text(2005, 14, 'Men')
    elif cat_index == 5:
        ax.text(2005, 87, 'Men')
        ax.text(2005, 7, 'Women')
        ax.tick_params(labelbottom='on')

# Generating the second column
# Here I generate line charts for both male and female percentages for every degree in the Liberal Arts category
for sp in range(1,16,3):
    cat_index = int(sp/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[cat_index]], c='blue', label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[cat_index]], c='orange', label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[cat_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    ax.set_yticks([0,100])
    ax.axhline(50, c=(207/255, 207/255, 207/255), alpha=0.3)
    
    if cat_index == 0:
        ax.text(2005, 75, 'Women')
        ax.text(2005, 20, 'Men')
    elif cat_index == 4:
        ax.text(1972, 65, 'Men')
        ax.text(1972, 30, 'Women')
        ax.tick_params(labelbottom='on')
        
# Generating the 3rd Column
# Here I generate line charts for both male and female percentages for every degree in the Other category
for sp in range(2,18,3):
    cat_index = int(sp/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[cat_index]], c='blue', label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[cat_index]], c='orange', label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[cat_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    ax.set_yticks([0,100])
    ax.axhline(50, c=(207/255, 207/255, 207/255), alpha=0.3)
    
    if cat_index == 0:
        ax.text(2005, 88, 'Women')
        ax.text(2005, 20, 'Men')
    elif cat_index == 5:
        ax.text(2005, 35, 'Women')
        ax.text(2005, 60, 'Men')
        ax.tick_params(labelbottom='on')

# I am exporting all graphs as a PNG file, so that it could be used in Word Documents, Powerpoint presentations and etc.
plt.savefig('gender_degrees.png')
plt.show()
```
